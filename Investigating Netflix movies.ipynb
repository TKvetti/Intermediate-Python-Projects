1. Load the CSV file and store as netflix_df.

```python
# import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

#load csv file and store as netflix_df
netflix_df = pd.read_csv("netflix_data.csv", index_col = 0)
netflix_df
```

2. Filter the data to remove TV shows and store as netflix_subset.

```python
# filter to remove TVshows and store as netflix_subset
#(select only movies from under the column type)

netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
netflix_subset
```

3. Investigate and subset the Netflix movie data, keeping only the columns "title", "country", "genre", "release_year", "duration", and saving this into a new DataFrame called netflix_movies.

```python
# select required columns for analysis from netflix_subset and store as netflix_movies

netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
netflix_movies
```

4. Filter netflix_movies to find the movies that are strictly shorter than 60 minutes, saving the resulting DataFrame as short_movies; inspect the result to find possible contributing factors.

```python
#filter for movies shorter than 60 mins and store as short_movies

short_movies = netflix_movies[netflix_movies["duration"] < 60]
short_movies
```

5.   a. Using a for loop and if/elif statements, iterate through the rows of netflix_movies and assign colors of your choice to four genre groups ("Children", "Documentaries", "Stand-Up", and "Other" for everything else). Save the results in a colors list.

   b.  Initialize a matplotlib figure object called fig and create a scatter plot for movie duration by release year using the colors list to color the points and using the labels "Release year" for the x-axis, "Duration (min)" for the y-axis, and the title "Movie Duration by Year of Release". 

```python
#assign colors for genre
colors = []
for lab, row in netflix_movies.iterrows():
    if "Children" in row["genre"]:
        colors.append("orange")
    elif "Documentaries" in row["genre"]:
        colors.append("green")
    elif "Stand-Up" in row["genre"]:
        colors.append("blue")
    else:
        colors.append("black")
```

```python

#create mtplotlib figure object using plt.figure

#use the syntax figure(figsize =(width, length))

#given figure size is 12 width and 8 length

fig = plt.figure(figsize= (12, 8))

plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')

#scatterplot
plt.scatter(netflix_movies["release_year"], netflix_movies["duration"], c = colors)
plt.show()
```
